```python
# Required imports
import matplotlib.pyplot as plt
import numpy as np
from cool_maps.plot import set_mapping_library, add_bathymetry, create
```

```python
# Define extent and bathymetry parameters
extent = [-99, -78, 18, 33]

# Common settings
levels = [-8000, -1000, -100, 0]
figsize = (20, 16)
```

```python
from cool_maps.download import get_bathymetry

# Get bathymetry data
bathy = get_bathymetry(extent)
bathy

# Save bathymetry data into variables
lon = bathy['longitude']
lat = bathy['latitude']
elevation = bathy['z']
```

```python
import cartopy.crs as ccrs
# Generate Cartopy map
set_mapping_library('cartopy')
fig, ax = create(extent, figsize=figsize, ticks=False, proj=ccrs.Mercator(), bathymetry=False)
add_bathymetry(ax, lon, lat, elevation, levels=levels, method='tri_color', add_legend=True)
ax.set_title('Cartopy Bathymetry')
plt.show()

# Generate Basemap map
set_mapping_library('basemap')
fig, ax = create(extent, figsize=figsize, ticks=False, proj=None, bathymetry=False)
add_bathymetry(ax, lon, lat, elevation, levels=levels, method='tri_color', add_legend=False)
ax.set_title('Basemap Bathymetry')
plt.show()
```

```python
# Generate Basemap map
set_mapping_library('basemap')
fig, ax = create(extent, figsize=figsize, ticks=False, proj=None, bathymetry=False)
add_bathymetry(ax, lon, lat, elevation, levels=levels, method='tri_color', add_legend=False)
ax.set_title('Basemap Bathymetry')
plt.show()
```
